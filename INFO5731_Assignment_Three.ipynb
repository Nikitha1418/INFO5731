```
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from textblob import Word
import pandas as pd
import nltk
```

```
def special_characters_count(given_input):
  number_of_special_characters = 0
  for i in range(len(given_input)):
    if not((given_input[i].isalpha()) and (given_input[i].isdigit())):
        number_of_special_characters = number_of_special_characters + 1
  return number_of_special_characters
```

```
#Removal of special characters:

input_data['Removal of special characters'] = input_data['Reviews'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))

#Removal of punctuations :

input_data['Punctuation'] = input_data['Removal of special characters'].str.replace('[^\w\s]','')

#Removal of numbers:

input_data['Removal of numbers'] = input_data['Punctuation'].str.replace('\d+', '')

#Lower case all texts:

input_data['Lower Case words'] = input_data['Removal of numbers'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#Removing stopwords by using stopwordslist

stop = stopwords.words('english')
input_data['removed stopwords'] = input_data['Lower Case words'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```
import nltk
from textblob import TextBlob
nltk.download('punkt')
input_data['tokenized data'] = input_data['removed stopwords'].apply(lambda x: TextBlob(x).words)
```

```
from nltk.stem import PorterStemmer
st = PorterStemmer()
input_data['Stemming'] = input_data['tokenized data'].apply(lambda x: " ".join([st.stem(word) for word in x]))
```

```
from textblob import Word
import nltk
nltk.download('wordnet')
input_data['Lemmatization'] = input_data['Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
input_data
```

```
import itertools
import collections
from nltk import ngrams
words = []
for i in input_data['Lemmatization']:
  words.append(nltk.tokenize.word_tokenize(i))
data = [x for x in words if x != []]
iter_values = list(itertools.chain.from_iterable(data))
distributions = nltk.FreqDist(nltk.trigrams(iter_values))
distributions
```

```
distributions = nltk.FreqDist(nltk.bigrams(iter_values))
distributions_dict = dict(distributions)
for i in distributions_dict:
  value = str(i) + '->' + str(distributions_dict[i]/iter_values.count(i[1]))
  print(value)
```

```
def remove_duplicates(ip):
  return set(ip)
```

```
data_list = input_data["Lower Case words"].values.tolist()
words = [j for i in data_list for j in i.split(" ")]
words = remove_duplicates(words)
# print(words)
dataframe1 = pd.DataFrame(words,columns=["words"])
count = 1
import math
def tf_idf(word, sentence):
  value = sentence.count(word)
  size = len(sentence)
  if(value!=0):
    return (value/size)*(math.log(size/value,10))
  else:
    return 0
  return 


for data in data_list:
  dataframe1[count] = dataframe1["words"].apply(lambda i : tf_idf(i, data))
  count=count+1

dataframe1
```

```
import numpy as np
cosine = []
X = "I was looking to replace a 7 year old laptop that seen better days...I looked for months on and off . I add this one to my list because it matched the dimensions of the old laptop with some upgrades and minus other stuff i don't need anymore.  Just like everyone else I was pleased with a new laptop. I had an issue with with Tag # not matching what the software was showing. I called Dell to see what the issue was. The Agent informed me that the laptop was upgrade by an authorized vendor , they forgot to update the Tag # to show it's been upgraded. The Agent then proceeded to tell me the upgrades bumped the cost of the unit from 3 00 to $400 higher than the base unit. I went on Dells website to price it out and he was right ! I got way more than what I paid for it !!!! I have no complaints with the unit aside from the battery life."
for line in input_data['Lemmatization']:
  list_1 = list_2 = []
  v = 0
  Tokens_X = nltk.tokenize.word_tokenize(X)
  Tokens_Y = nltk.tokenize.word_tokenize(line)
  stop_words = nltk.corpus.stopwords.words('english') 
  X_after_stop_words_removal = {word for word in Tokens_X if not word in stop_words}
  Y_after_stop_words_removal = {word for word in Tokens_Y if not word in stop_words}
  vector = X_after_stop_words_removal.union(Y_after_stop_words_removal)  
  for w in vector: 
    if w in X_after_stop_words_removal: list_1.append(1)
    else: list_1.append(0) 
    if w in X_after_stop_words_removal: list_2.append(1) 
    else: list_2.append(0) 
  for i in range(len(vector)): 
    v+= list_1[i]*list_2[i] 
  cosine_value = v / float((sum(list_1)*sum(list_2))**0.5)
  cosine.append(cosine_value)
reviews_id = np.arange(1, 112)
c_df = pd.DataFrame(list(zip(reviews_id, cosine)), columns = ['Id', 'Similarity'])
c_df
```
