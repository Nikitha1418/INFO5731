(1) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
import array as arry

def finding_duplicates(a): 
    _size = len(a) 
    Duplicate_array = [] 
    for i in range(_size): 
        k = i + 1
        for j in range(k, _size): 
            if a[i] == a[j] and a[i] not in Duplicate_array: 
                Duplicate_array.append(a[i]) 
    if len(Duplicate_array) != 0:
      return(Duplicate_array)
    else:
      return -1
             
input = arry.array('i', [ 1,2,2,5,9,9,8,10,5,6,])

print(finding_duplicates(input))
```

(2) Write a Python program to select all the Sundays of a specified year. (4 points)



```python
from datetime import date, timedelta
def year_sunday(input_year):
  de = date(input_year,1,1)
  de += timedelta(days = 6 - de.weekday())
  while de.year == input_year:
    yield de
    de += timedelta(days = 7)

for s in year_sunday(2019):
  print(s)

```
